```python
import scipy.io
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import ticker

# Define the font properties
font = {'family': 'Times New Roman', 'weight': 'bold', 'size': 24}

# Load the .mat file for the first plot
mat = scipy.io.loadmat(r'C:\Users\aes21009\Downloads\VRMS_GFM_F.mat')
varname = list(mat.keys())[3]  # Skipping the meta-information keys
data1 = mat[varname]
data1 = data1[:, 1]  # Only use the 2nd column (0-indexed)
data1 = data1 / 2521 # Normalize by 2521
time1 = np.arange(len(data1)) # Create time array

# Load the .mat file for the second plot
mat = scipy.io.loadmat(r'C:\Users\aes21009\Downloads\VRMS_GFM_F1.mat')
varname = list(mat.keys())[3]  # Skipping the meta-information keys
data2 = mat[varname]
data2 = data2[:, 1]  # Only use the 2nd column (0-indexed)
data2 = data2 / 2401.78 # Normalize by 2401.78
time2 = np.arange(len(data2)) # Create time array

# Plot the data
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(time1, data1, label='SM', linewidth=3)
ax.plot(time2, data2, label='GFM', linestyle='--', linewidth=3)

ax.grid(True)
ax.set_xlabel('Time (µs)', fontdict=font)
ax.set_ylabel('Voltage (p.u.)', fontdict=font)

# Add a vertical line to the plot
ax.axvline(1.01e6, color='red', linestyle=':', linewidth=3, label='Fault')

# Set x-limits
ax.set_xlim([800000, max(time1)])
ax.get_xaxis().get_major_formatter().set_scientific(False)
# Set font style for x and y ticks
for label in ax.get_xticklabels():
    label.set_fontname('Times New Roman')
    label.set_fontweight('bold')
    label.set_fontsize(24)
for label in ax.get_yticklabels():
    label.set_fontname('Times New Roman')
    label.set_fontweight('bold')
    label.set_fontsize(24)
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:0.1f}e6'.format(x*1e-6)))


# Set font size for legend
legend = ax.legend()
for text in legend.get_texts():
    text.set_fontsize(24)
    text.set_fontname('Times New Roman')
    text.set_fontweight('bold')
    
# Save the figure
plt.savefig('GFM_F.pdf', format='pdf')

plt.show()
```
